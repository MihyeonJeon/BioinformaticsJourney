### **1. The Series Data Structure**

* The series is one of the core data structures in pandas.
* It's like a cross between a list and a dictionary.

```python
import pandas as pd
```

```python
numbers = [1, 2, 3]
pd.Series(numbers)
```

```python
students = ['Alice', 'Jack', 'Molly']
pd.Series(students)
```

* In Python, there is the none type to indicate a lack of data.
*  What if we want to have a typed list?
* Pandas does some type conversion.


```python
students = ['Alice', 'Jack', None]
# And lets convert this to a series
pd.Series(students)
```

* If we create a list of numbers, integers or floats, and put int the None type, pandas automatically converts this to a special floating point value designated as NaN which stands for **"Not a Number"**

```python
# So lets create a list with a None value in it
numbers = [1, 2, None]
# And turn that into a series
pd.Series(numbers)
```

*  First, NaN is a different value.
* Second, pandas set the dtype of this series to floating point numbers instead of object or ints.
---
* NaN is **NOT** equivalent to None.
* Actually, we can't do an equality test of NaN to itself.
* Instead, we need to use special functions to test for the presence of not a number.
* The  meaning of NaN is similar to None, but it's a numeric value and treated differently.

```python
import numpy as np
# And lets compare it to None
np.nan == None
```

```python
np.nan == np.nan
```

```python
np.isnan(np.nan)
```

* The ipython interpreter has something called magic functions begin with a percentage sign. 
* Here, we're actually going to use what's called a cellular magic function.

```python
%%timeit -n 100
total = 0
for number in numbers:
    total+=number

total/len(numbers)
```

```python
%%timeit -n 100
total = np.sum(numbers)
total/len(numbers)
```

* This is a pretty shocking difference in the speed and demonstrates why one should be aware of parallel computing features and start thinking in functional programming terms.
---
* A Related feature in pandas and nummy is called broadcasting. 
* With broadcasting, you can apply an operation to every value in the series, changing the series. 
* For instance, if we wanted to increase every random variable by 2, we could do so quickly using the += operator directly on the Series object. 


```python
numbers.head()
```

```python
numbers+=2
numbers.head()
```

```python
for label, value in numbers.iteritems():
    # now for the item which is returned, lets call set_value()
    numbers.set_value(label, value+2)
# And we can check the result of this computation
numbers.head()
```

```python
pd.__version__
```

```python
%%timeit -n 10
# we'll create a blank new series of items to deal with
s = pd.Series(np.random.randint(0,1000,1000))
# And we'll just rewrite our loop from above.
for label, value in s.iteritems():
    s.loc[label]= value+2
```

```python
%%timeit -n 10
# We need to recreate a series
s = pd.Series(np.random.randint(0,1000,1000))
# And we just broadcast with +=
s+=2
```

* It's easy to delete data in Series and DataFrames, and we can use the drop function to do so.
* This function takes a single parameter, which is the index or row label, to drop. 
* The drop function doesn't change the DataFrame by default! 
* Instead, the drop function returns to you a copy of the DataFrame with the given rows removed.


```python
df.drop('school1')
```

```python
df
```

* Drop has two interesting optional parameters. The first is called inplace, and if it's set to true, the DataFrame will be updated in place, instead of a copy being returned. 
* The second parameter is the axes, which should be dropped. By default, this value is 0, indicating the row axis. But you could change it to 1 if you want to drop a column.


```python
# For example, lets make a copy of a DataFrame using .copy()
copy_df = df.copy()
# Now lets drop the name column in this copy
copy_df.drop("Name", inplace=True, axis=1)
copy_df
```

* There is a second way to drop a column, and that's directly through the use of the indexing operator, using the del keyword. 
* This way of dropping data, however, takes immediate effect on the DataFrame and does not return a view.

```python
del copy_df['Class']
copy_df
```

```python
df['ClassRanking'] = None
df
```

###  **3. Querying a DataFrame**

* The first step in querying DataFrame is to understand **Boolean masking.** 
* Boolean masking is the heart of fast and efficient querying in numpy and pandas, and its analogous to bit masking used in other areas of computational science. 

```python
df = pd.read_csv('/content/drive/MyDrive/etc/Admission_Predict.csv', index_col=0)
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
admit_mask=df['chance of admit'] > 0.7
admit_mask
```

* The result of broadcasting a comparison operator is a Boolean mask - true or false values depending upon the results of the comparison.
* You can just lay it on top of the data to "hide" the data you don't want, which is represented by all of the False values. 

```python
df.where(admit_mask).head()
```

* All of the rows which did not meet the condition have NaN data instead, but these rows were not dropped from our dataset. 
* The next step is, if we don't want the NaN data, we use the dropna() function

```python
df.where(admit_mask).dropna().head()
```

* Despite being really handy, where() isn't actually used that often. Instead, the pandas devs created a shorthand syntax which combines where() and dropna(), doing both at once. 
* In typical fashion, the just overloaded the indexing operator to do this!

```python
df[df['chance of admit'] > 0.7].head()
```

```python
df["gre score"].head()
```

```python
df[["gre score","toefl score"]].head()
```

```python
df[df["gre score"]>320].head()
```

```python
# (df['chance of admit'] > 0.7) and (df['chance of admit'] < 0.9)
(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
```

* One thing to watch out for is order of operations! A common error for new pandas users is to try and do boolean comparisons using the & operator but not putting **parentheses** around the individual terms you are interested in.

```python
df['chance of admit'] > 0.7 & df['chance of admit'] < 0.9
```

```python
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```python
df['chance of admit'].gt(0.7).lt(0.9)
```

### **6. Example : Manipulating DataFrame**

```python
df=pd.read_csv("/content/drive/MyDrive/etc/presidents.csv")
df.head()
```

```python
df["First"]=df['President']
df["First"]=df["First"].replace("[ ].*", "", regex=True)
df.head()
```

* It works but slow. We can use apply() function.

```python
del(df["First"])

def splitname(row):
    # The row is a single Series object which is a single row indexed by column values
    # Let's extract the firstname and create a new entry in the series
    row['First']=row['President'].split(" ")[0]
    # Let's do the same with the last word in the string
    row['Last']=row['President'].split(" ")[-1]
    # Now we just return the row and the pandas .apply() will take of merging them back into a DataFrame
    return row

# Now if we apply this to the dataframe indicating we want to apply it across columns
df=df.apply(splitname, axis='columns')
df.head()
```

* There are better ways to separate first and last name.

```python
del(df['First'])
del(df['Last'])

pattern="(^[\w]*)(?:.* )([\w]*$)"

df["President"].str.extract(pattern).head()
```

```python
pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names=df["President"].str.extract(pattern).head()
names
```

```python
# And we can just copy these into our main dataframe if we want to
df["First"]=names["First"]
df["Last"]=names["Last"]
df.head()
```

* Let's clean up 'Born' column!

```python
df["Born"]=df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df["Born"].head()
```

* Update the column to the data type.

```python
df["Born"] = pd.to_datetime(df["Born"])
df["Born"].head()
```
