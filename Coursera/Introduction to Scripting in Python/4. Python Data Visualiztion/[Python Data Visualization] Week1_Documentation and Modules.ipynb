### 2. Draw a county-level map of the USA in matplotlib

```python
"""
Week 1 practice project template for Python Data Visualization
Load a county-level PNG map of the USA and draw it using matplotlib
"""

PATH = '/content/drive/MyDrive/etc/'

import matplotlib.pyplot as plt

# Houston location

USA_SVG_SIZE = [555, 352]
HOUSTON_POS = [302, 280]

def draw_USA_map(map_name):
    """
    Given the name of a PNG map of the USA (specified as a string),
    draw this map using matplotlib
    """
     
    # Load map image, note that using 'rb'option in open() is critical since png files are binary
    with open(map_name, 'rb') as map_file:
        map_img = plt.imread(map_file)

    #  Get dimensions of USA map image
    ypix, xpix, bands = map_img.shape
    print(xpix, ypix, bands)

    # Plot USA map
    image = plt.imshow(map_img)
    
    # Plot green scatter point in center of map
    plt.scatter(x = xpix / 2, y = ypix / 2 , s = 100, c = 'Green')

    # Plot red scatter point on Houston, Tx - include code that rescale coordinates for larger PNG files
    plt.scatter(x = HOUSTON_POS[0] * xpix / USA_SVG_SIZE[0] , y = HOUSTON_POS[1] * ypix / USA_SVG_SIZE[1], s = 100, c= 'Red')
    plt.show()

```

```python
draw_USA_map(PATH+"USA_Counties_555x352.png")
draw_USA_map(PATH+"USA_Counties_1000x634.png")   
```

* height , width , channel = image.shape
* channel means the informationi of color in the image.
* The values of xpix and ypix are needed to find the center of the plot, but more importantly to resize the plot.
